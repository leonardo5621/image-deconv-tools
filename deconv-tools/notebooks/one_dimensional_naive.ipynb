```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

```python
## TEST WITH A TRIANGULAR SHAPED CURVE
L = 128
t = np.linspace(0, 1, L)
triangle = signal.sawtooth(2 * np.pi * t, 0.5) + 1
plt.plot(t, triangle)
```

```python
## SMOOTHING KERNEL
h = np.ones(5)/5

## GAUSSIAN NOISE
noise = 0.02*np.random.normal(loc=0, scale=1, size=t.shape)
```

```python
## SIGNAL TO NOISE RATIO
y = np.convolve(triangle, h, mode='same') + noise
plt.plot(t, y)

SNR = 10*np.log10(np.sum(np.square(y))/np.sum(np.square(noise)))
print(SNR)
```

```python
## FOURIER TRANSFORM

Y = np.fft.fft(y)
frequency = np.fft.fftfreq(t.shape[-1])
plt.plot(frequency, np.abs(Y))
```

```python
## REVERTING THE FILTER EFFECT
H = np.fft.fft(h, n=L)
plt.plot(frequency, np.abs(H))
X = np.divide(Y, H)
x = np.fft.ifft(X)
```

```python
plt.plot(t, x.real)
error = np.sum(np.square(x.real - triangle))
print(error)
```
